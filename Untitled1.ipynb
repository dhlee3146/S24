```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import time

# 텍스트
text = "Moving Text!"

# 초기 위치
x = np.linspace(-3, 3, len(text))
y = np.zeros_like(x)

# 시각화 설정
fig, ax = plt.subplots(figsize=(8, 1))
plt.axis('off')  # 플롯 축을 제거하여 텍스트만 표시
txt_objects = [ax.text(xi, yi, char, fontsize=100) for xi, yi, char in zip(x, y, text)]

# 애니메이션
for _ in range(50):  # 애니메이션 반복 횟수
    for txt_obj, xi in zip(txt_objects, x):
        xi += np.random.uniform(-0.5, 0.5)  # 각 글자마다 랜덤한 속도로 좌우로 이동
        txt_obj.set_position((xi, 0))  # y축은 고정
    clear_output(wait=True)
    display(fig)
    time.sleep(0.1)  # 속도 조절

plt.close()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output
import time

# 텍스트
text = "go right!"

# 초기 위치
word_length = len(text)
x_initial = np.linspace(-1, 1, word_length)  # 초기 x 좌표 설정

# 시각화 설정
fig, ax = plt.subplots(figsize=(8, 1))
plt.axis('off')  # 플롯 축을 제거하여 텍스트만 표시
txt_objects = [ax.text(xi, 0, char, fontsize=50) for xi, char in zip(x_initial, text)]

# 애니메이션
for _ in range(50):  # 애니메이션 반복 횟수
    for txt_obj in txt_objects:
        txt_obj.set_position((txt_obj.get_position()[0] + 0.1, 0))  # 모든 글자를 오른쪽으로 이동
    clear_output(wait=True)
    display(fig)
    time.sleep(0.1)  # 속도 조절

plt.close()
```
